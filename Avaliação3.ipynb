```python
#Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfPrincipal = pd.read_csv('https://raw.githubusercontent.com/EuNetu/DataSetNetflix/main/netflix_titles.csv')
```

```python
#dfPrincipal
```

```python
#Mostra os filmes e series lançados no ano de 2021
df2021 = dfPrincipal[dfPrincipal['release_year'] == 2021]

#Mostra os filmes e series lançados no ano de 2020
df2020 = dfPrincipal[dfPrincipal['release_year'] == 2020]

#Mostra os filmes e series lançados no de 2019
df2019 = dfPrincipal[dfPrincipal['release_year'] == 2019]

#Mostra os filmes e series lançados no de 2018
df2018 = dfPrincipal[dfPrincipal['release_year'] == 2018]

#Mostra apenas filmes lançados no ano de 2021
filme2021 = df2021[df2021['type'] == 'Movie']

#Mostra apenas filmes lançados no ano de 2020
filme2020 = df2020[df2020['type'] == 'Movie']

##Mostra apenas filmes lançados no ano de 2019
filme2019 = df2019[df2019['type'] == 'Movie']

##Mostra apenas filmes lançados no ano de 2018
filme2018 = df2018[df2018['type'] == 'Movie']
```

```python
#filme2021
#filme2020
#filme2019
#filme2018
```

```python
#retira algumas colunas do df
filme2021Fil = filme2021.drop(columns=['rating', 'cast', 'description', 'show_id', 'type', 'date_added'])

#retira algumas colunas do df
filme2020Fil = filme2020.drop(columns=['rating', 'cast', 'description', 'show_id', 'type', 'date_added'])

#retira algumas colunas do df
filme2019Fil = filme2019.drop(columns=['rating', 'cast', 'description', 'show_id', 'type', 'date_added'])

#retira algumas colunas do df
filme2018Fil = filme2018.drop(columns=['rating', 'cast', 'description', 'show_id', 'type', 'date_added'])
```

```python
#filme2021Fil
#filme2020Fil
#filme2019Fil
#filme2018Fil
```

```python
#Mudando os nomes das colunas
filme2021Rename = filme2021Fil.rename(columns={'title': 'título', 'director': 'Diretor', 'country': 'País', 'release_year': 'Ano de Lançamento', 'duration': 'Duração', 'listed_in': 'Gênero'})

#Mudando os nomes das colunas
filme2020Rename = filme2020Fil.rename(columns={'title': 'título', 'director': 'Diretor', 'country': 'País', 'release_year': 'Ano de Lançamento', 'duration': 'Duração', 'listed_in': 'Gênero'})

#Mudando os nomes das colunas
filme2019Rename = filme2019Fil.rename(columns={'title': 'título', 'director': 'Diretor', 'country': 'País', 'release_year': 'Ano de Lançamento', 'duration': 'Duração', 'listed_in': 'Gênero'})

#Mudando os nomes das colunas
filme2018Rename = filme2018Fil.rename(columns={'title': 'título', 'director': 'Diretor', 'country': 'País', 'release_year': 'Ano de Lançamento', 'duration': 'Duração', 'listed_in': 'Gênero'})
```

```python
#filme2021Rename
#filme2019Rename
#filme2020Rename
#filme2020Rename
```

```python
#preenchendo dados ausentes com -
#DFfinal
DFfilme2021 = filme2021Rename.fillna(value='-')
DFfilme2020 = filme2020Rename.fillna(value='-')
DFfilme2019 = filme2019Rename.fillna(value='-')
DFfilme2018 = filme2018Rename.fillna(value='-')
```

```python
#DFfilme2021
#DFfilme2020
#DFfilme2019
#DFfilme2018
```

```python
#Filtra filmes brasileiros de 2021
brasil2021 = DFfilme2021[(DFfilme2021['País'] == 'Brazil')]

#Filtra filmes brasileiros de 2021
brasil2020 = DFfilme2020[(DFfilme2020['País'] == 'Brazil')]

#Filtra filmes brasileiros de 2021
brasil2019 = DFfilme2019[(DFfilme2019['País'] == 'Brazil')]

#Filtra filmes brasileiros de 2021
brasil2018 = DFfilme2018[(DFfilme2018['País'] == 'Brazil')]
```

```python
#Filtra filmes brasileiros de 2021
EUA2021 = DFfilme2021[(DFfilme2021['País'] == 'United States')]

#Filtra filmes brasileiros de 2021
EUA2020 = DFfilme2020[(DFfilme2020['País'] == 'United States')]

#Filtra filmes brasileiros de 2021
EUA2019 = DFfilme2019[(DFfilme2019['País'] == 'United States')]

#Filtra filmes brasileiros de 2021
EUA2018 = DFfilme2018[(DFfilme2018['País'] == 'United States')]
```

```python
#Concatenando filmes brasileiros de 2018 - 2019
resultBra18_19 = pd.concat([brasil2018, brasil2019])
resultBra18_19
```

```python
#Concatenando filmes Estadunidenses de 2018 - 2019
resultEUA18_19 = pd.concat([EUA2018, EUA2019])
resultEUA18_19
```

```python
#Concatenando filmes Brasileiros e Estadunidenses de 2018 - 2019
result18_19 = pd.concat([resultBra18_19, resultEUA18_19])
result18_19
```

```python
#Concatenando filmes brasileiros de 2020 - 2021
resultBra20_21 = pd.concat([brasil2020, brasil2021])
resultBra20_21
```

```python
#concatenando filmes dos Estados Unidos de 2020 - 2021
resultEUA20_21 = pd.concat([EUA2020, EUA2021])
resultEUA20_21
```

```python
#Concatenando filmes brasileiros e Estadunidenses de 2020 - 2021
result20_21 = pd.concat([resultEUA20_21, resultBra20_21])
result20_21
```

Comparação entre Brasil e Estados Unidos, com base nos filmes lançados nos anos de 2020 e 2021.

```python
#Concatenando filmes de 2018 - 2019 - 2020 - 2021
resultGeral = pd.concat([DFfilme2021, DFfilme2020, DFfilme2019, DFfilme2018])
resultGeral
```

Comparação entre a quantidade de filmes lançados nos anos de 2018, 2019, 2020 e 2021.

Gráfico 3:

```python
#Gráfico Pizza
resultGeral['Ano de Lançamento'].value_counts().plot.pie(autopct='%.0f%%')
```

Visualização em nuvem de palvras, de gêneros lançados nos anos de 2018, 2019, 2020 e 2021.

Imagem 1:

```python
from wordcloud import WordCloud, STOPWORDS
generos = ' '.join(resultGeral['Gênero'].dropna())

# gerando uma wordcloud dos genêros dos filmes
wordcloudAutores = WordCloud(background_color="black", width=1600, height=800).generate(generos)

# Plotando a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.set_axis_off()
plt.imshow(wordcloudAutores);
```

Visualização de nuvem de palavras, dos países com lançamentos nos anos de 2018, 2019, 2020 e 2021.

Imagem 2:

```python
from wordcloud import WordCloud, STOPWORDS
generos = ' '.join(resultGeral['País'].dropna())

# gerando uma wordcloud dos genêros dos filmes
wordcloudAutores = WordCloud(background_color="black", width=1600, height=800).generate(generos)

# Plotando a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.set_axis_off()
plt.imshow(wordcloudAutores);
```
